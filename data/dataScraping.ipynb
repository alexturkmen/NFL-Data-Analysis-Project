```python
import pandas as pd
from pathlib import Path
import numpy as np
from bs4 import BeautifulSoup
import requests
```

# Summary 
This file scrapes player salary data from pro-football-reference.com using the current webpage and saves from the previous 5 years on web.archive.

It also scrapes fantasy data for those years, but this information was excluded from the project as extraneous as we found a separate json dataset that had all NFL players (all-time). 

## 2020 Player Salary Data - Pro Football Reference

```python
url = 'https://www.pro-football-reference.com/players/salary.htm'
df2020 = pd.read_html(url)[0]
writepath = Path.cwd() / 'Old' / 'salaryData2020.csv'
df2020['year'] = '2020'
df2020.rename(columns = {'Salary':'Cap Hit'}, inplace=True)
df2020.drop(['Rk'], axis=1, inplace=True)
df2020.to_csv (writepath, index = False, header=True)
df2020.head()
```
